```python
import warnings
warnings.filterwarnings("ignore")
import tensorflow as tf
print(tf.__version__)
```

```python
import keras
mnist = keras.datasets.fashion_mnist
```

```python
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(training_images[0])

print(training_labels[0])
print(training_images[0])
```

```python
#normalizing data between 0 and 1
training_images  = training_images / 255.0
test_images = test_images / 255.0
```

```python

import keras
from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.optimizers import Adam



class myCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        #print("At epoch end:")
        #print("logged accuracy =>",logs.get('accuracy'))
        if(logs.get('accuracy')>0.91):
          print("\nReached 92% accuracy so cancelling training!")
          self.model.stop_training = True

callbacks = myCallback()

model = Sequential([
                    Flatten(input_shape=(28, 28)),
                    #works fine until you hit the law of diminishing returns, in other words give more accurate result but trains slower
                    #Dense(1024, activation=tf.nn.relu)
                    
                    Dense(128, activation=tf.nn.relu), 
                    #For far more complex data (including color images to be classified as flowers that you'll see in the next lesson), extra layers are often necessary
                    #Dense(256, activation=tf.nn.relu)
                    
                    Dense(10, activation=tf.nn.softmax)])
```

```python

model.compile(optimizer = Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'],
             )
```

```python
model.fit(training_images, training_labels, epochs=30,callbacks=[callbacks])
```

```python
model.evaluate(test_images, test_labels)
```

```python
classifications = model.predict(test_images)
```
